```python
# A car salesman sells on an average 3 cars per week. [3 Marks]
# 3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
# 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
# 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]
```

```python
import numpy as np
from scipy import stats
lamda = 3 #average sales is 3 cars/week
n=np.arange(0,20)
p=stats.poisson.pmf(n,lamda)
print(p)
```

```python
pcum=stats.poisson.cdf(n,lamda)
print(pcum)
```

```python
print(' Probability that in a given week he will sell some cars is %1.4f ' % pcum[1])
```

```python
Car_Range=pcum[5]-pcum[1]
print(' Probability that in a given week he will sell 2 or more but less than 5 cars %1.4f' % Car_Range)
```

```python
# Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(n,pcum, 'o-')
plt.title('Poisson')
plt.xlabel('Number Cars sold/week')
plt.ylabel('Cum Probability of Cars sold/week')
plt.show()
```

```python
# Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed,
# marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the
# percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours
# independently place orders with the same bot. Answer the following questions. [3 Marks]
# 4.A. What is the probability that all three orders will be recognised correctly?
# 4.B. What is the probability that none of the three orders will be recognised correctly?
# 4.C. What is the probability that at least two of the three orders will be recognised correctly?"""
```

```python
from scipy import stats
import numpy as np
acc=0.868 # percentage of orders taken correctly
n=3 # bot and two friends
k=np.arange(0,4)
binom=stats.binom.pmf(k=k,n=n,p=acc)
binom
```

```python
print("Probability that all 3 orders will be recognised correctly is %1.4f" %binom[3])
```

```python
print("Probability that none of the 3 orders will be recognised correctly is %1.4f" %binom[0])
```

```python
cumbinom=stats.binom.cdf(k,n,acc)
cumbinom
```

```python
cumbinom1=1-cumbinom[1]
print("Probability that atleast 2 of the 3 orders will be recognised correctly is %1.4f" %cumbinom1)
```
